### **Tuples**



**Definations**: A tuple is a collection of different data types(Hetrogeneous) which is **ordered** and **unchangeable (immutable)**. Tuples are written with round brackets, (). Once a tuple is created, we cannot change its values. We cannot use add, insert, remove methods in a tuple because it is not modifiable (mutable). Unlike list, tuple has few methods. Methods related to tuples:

- tuple(): to create an empty tuple
- count(): to count the number of a specified item in a tuple
- index(): to find the index of a specified item in a tuple
- "+" operator: to join two or more tuples and to create a new tuple

```python
# Emplty tuple

empty_tuple = ()

print(empty_tuple)
print(type(empty_tuple))
print("-"*25)
# or using the tuple constructor
empty_tuple = tuple()
print(empty_tuple)
print(type(empty_tuple))
```

```python
# Tuple with items
my_tuple = (1, 2, 3, "Hello", [4, 5], (6, 7))
print(my_tuple)
print(type(my_tuple))
print(len(my_tuple))
```

```python
# Creating a tuple without parentheses
my_tuple = 1, 2, 3, "Hello"
print(my_tuple)
print(type(my_tuple))
print(len(my_tuple))
```

```python
# Tuple with single item
single_item_tuple = (42)
print(single_item_tuple)
print(type(single_item_tuple))
```

```python
# Tuple with single item
single_item_tuple = (42,)
print(single_item_tuple)
print(type(single_item_tuple))
```

```python
# Then we can use a list to change the tuple
my_tuple = (1, 2, 3, "Hello", [4, 5], (6, 7))
my_list = list(my_tuple)
my_list[0] = 10
my_tuple = tuple(my_list)
print(my_tuple)
```

```python
# Joining two tuples
fruits = ('banana', 'orange', 'mango', 'lemon')
vegetables = ('Tomato', 'Potato', 'Cabbage','Onion', 'Carrot')
fruits_and_vegetables = fruits + vegetables
print(fruits_and_vegetables)
```

```python
# Deleting a tuple (this will raise an error)
del fruits
print(fruits)
```

```python
# Joining two tuples
fruits = ('banana', 'orange', 'mango', 'lemon')
vegetables = ('Tomato', 'Potato', 'Cabbage','Onion', 'Carrot')
fruits_and_vegetables = fruits + vegetables
print(fruits_and_vegetables)
```

```python
# Iterating through a tuple
for i in range(len(fruits_and_vegetables)):
    print(f'At index {i}: {fruits_and_vegetables[i]}')
```

```python
# Iterating through a tuple with enumerate
for i, item in enumerate(fruits_and_vegetables):
    print(f'At index {i}: {item}')
    
```
