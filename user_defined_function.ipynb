### User Define Function

Python Functions are a block of statements that does a specific task. The idea is to put some commonly or repeatedly done task together and make a function so that instead of writing the same code again and again for different inputs, we can do the function calls to reuse code contained in it over and over again.

```python
# Create a python function that takes a string as input and returns the string in reverse order.
def reverse_string(s):
    return s[::-1]

print(reverse_string("Hello, World!"))
```

```python
# Create a function that takes a string as input and returns the number of vowels and consonants in the string.
def count_vowels_consonants(s):
    vowels = "aeiouAEIOU"
    vowel_count = 0
    consonant_count = 0
    for char in s:
        if char.isalpha():
            if char in vowels:
                vowel_count += 1
            else:
                consonant_count += 1
    return vowel_count, consonant_count

vowels, consonants = count_vowels_consonants("Hello, World!")
print(f"Vowels: {vowels}, Consonants: {consonants}")
```

```python
user_input = input("Enter a string: ")
vowels, consonants = count_vowels_consonants(user_input)
print("You entered:", user_input)
print(f"Vowels: {vowels}, Consonants: {consonants}")
```

```python
# Create a function to check the entered number is Even or ODD

def check_odd_even(x: int):
    
    if x % 2 == 0:
        print("Entered number is Even")
    else:
        print("Entered number is ODD")
        
        
check_odd_even(5)
```

```python
# Variable length argument functions

# function defination
def fun(*args):
    for arg in args:
        print(arg)

# function calling
fun("Python", "Java", "C++")
```

```python
# Create a fuction that return the multication of two numbers

def multiplication(num1, num2):
    
    return num1*num2

multiplication(5,7)
```

```python
# Function to create a table

def table_x(x):
    for iter in range(1,11):
        print(x ," x ", iter, " = ", x*iter)
        
table_x(5)
```
