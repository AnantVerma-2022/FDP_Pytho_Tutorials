**Iterating through a String**

If we want to get individual characters of a string one by one, we can use a for loop.

```python
# lets suppose i have a string "Python" i want to print all the character one by one

text = "Python"
print(text[0])
print(text[1])
print(text[2])
print(text[3])
print(text[4])
print(text[5])
```

 - Think you have here a small string and that's why abled to manage the same but in real time you will have long sequence. Hence we need a sophosticated way to deal with the situation.

 Hence, the Control Statements are in python.
 - For
 - while

```python
text = "Python"

for char in text:
    print(char)
```

- In each iteration, the value of character is each individual character from the text string.

```python
# Count the occurrence of a substring in a string

text = "Python Programming is fun. Python is easy to learn."

count = 0
for i in text:
    if i in "aieouAEIOU":
        count += 1
    else:
        continue
print(count)
```

```python
# Create a user defined function to count the number of vowels in a string

def count_vowels(text):
    count = 0
    for i in text:
        if i in "aieouAEIOU":
            count += 1
        else:
            continue
    return count

count = count_vowels(text)
print(count)
```

```python
# Create a user funtion to check whether the user input is Palindrome or not.

def check_palindrome(text):
    
    if text[::-1] == text:
        print("The user input is palindrome")
    else:
        print("The user input is not a palindrome")
        
check_palindrome("madam")
```
