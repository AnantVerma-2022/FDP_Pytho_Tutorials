### **Dictionaries**

A dictionary is a collection of unordered, modifiable(mutable) paired (key: value) data type.

```python
# Create a dictionary - empty dictionary

myDictionary = {}
print(myDictionary)
print(type(myDictionary))

print('---' * 10)
myDictionary = dict()
print(myDictionary)
print(type(myDictionary))
```

```python
# Dictionary with some key-value pairs
myDictionary = {
    'name': 'John',
    'age': 30,
    'city': 'New York'
}
print(myDictionary)
print(type(myDictionary))
```

```python
# Print number of key-value pairs in the dictionary
print(len(myDictionary))
```

```python
# another example of a dictionary
person = {
    'first_name':'Anant',
    'last_name':'Verma',
    'age':32,
    'country':'India',
    'is_married':True,
    'skills':['Python', 'C++', 'Data Science', 'Machine Learning', 'Deep Learning'],
    'address':{
        'street':'Noida 137',
        'zipcode':'201301'
    }
    }
print(person)
print(len(person))
```

```python
# Accessing values in a dictionary using keys
print(person['first_name'])
```

```python
# Print address of the person
print(person['address'])
```

```python
# Accessing values in a dictionary using get() method
print(person.get('first_name'))
print(person.get('address'))
```

```python
# Add a new key-value pair to the dictionary
person['phone'] = '123-456-7890'
print(person)
```

```python
# print skills of the person
print(person['skills'])
```

```python
# Now  I am adding a new skill to the skills list
person['skills'].append('SQL')
print(person['skills'])
```

```python
# Changing the value of a key
person['country'] = 'USA'
print(person['country'])
```

```python
# Check if a key exists in the dictionary
print('age' in person)
print('salary' in person)
```

```python
# Remove a key-value pair from the dictionary
person.pop('is_married')
print(person)
```

```python
person.popitem()
```

```python
# Get list of keys
print(person.keys())
```

```python
# Get list of values
print(person.values())
```

```python
# Get lits of tuple of items
print(person.items())
```

```python
# Iterate through the dictionary
for key in person:
    print(key)
```

```python
# Iterate through the dictionary
for value in person.values():
    print(value)
```

```python
# Iteration 

for k,v in person.items():
    print(k)
```

```python
# Iteration 

for k,v in person.items():
    print(v)
```

```python
# Iteration 

for k,v in person.items():
    print(k,"--->", v)
```

```python
# Copy a dictionary

dict_copy = person.copy()
dict_copy
```
