**Question**: Slope is (m = y2-y1/x2-x1). Find the slope and Euclidean distance between point (2, 2) and point (6,10)

```python
# Points
x1, y1 = 2, 2
x2, y2 = 6, 10

# Slope
slope = (y2 - y1) / (x2 - x1)

# Euclidean Distance (manual square root using exponentiation)
distance = ((x2 - x1)**2 + (y2 - y1)**2) ** 0.5

print("Slope:", slope)
print("Euclidean Distance:", distance)
```

```python
# Using math module for square root
import math

distance_with_math = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
print("Euclidean Distance (using math.sqrt):", distance_with_math)

slope = (y2 - y1) / (x2 - x1)
print("Slope:", slope)
```

Get user input using input(“Enter your age: ”). If user is 18 or older, give feedback: You are old enough to drive. If below 18 give feedback to wait for the missing amount of years. Output:
