### Variables


Variables are used to store data in a computer’s memory. In many programming languages, it is recommended to use mnemonic variable names, which are names that are easy to remember and clearly related to the data they represent. A variable refers to a specific memory location where data is stored. When naming a variable, certain rules must be followed: it cannot begin with a number, include special characters, or contain hyphens. Although variables can have short names such as x, y, or z, using descriptive names like firstname, lastname, age, or country is strongly encouraged for better readability.

**Python Variable Naming Rules**

- A variable name must start with a letter or an underscore (_).

- A variable name cannot start with a number.

- A variable name may contain only alphanumeric characters and underscores (A–Z, a–z, 0–9, and _).

- Variable names are case-sensitive, meaning names such as firstname, Firstname, FirstName, and FIRSTNAME are treated as different variables.

**These are some examples:**
- firstname
- first_name
- last_name
- capital_city
- current_year_2026
- birth_year
- num1
- num2

```python
first_name = "Anant"
print(first_name)
```

```python
last_name = "Verma"
print(last_name)
```

```python
print("first_name :", first_name)
print("last_name :", last_name)
```

```python
print("first_name:", first_name, "last_name:", last_name)
```

```python
print('First name length:', len(first_name))
```

```python
print('Last name length: ', len(last_name))
```

```python
# Multiple variable assignment

first_name, last_name, country, age, is_cricket = 'Anant', 'Verma', 'India', 32, True

print(first_name, last_name, country, age, is_cricket)
print('First name:', first_name)
print('Last name: ', last_name)
print('Country: ', country)
print('Age: ', age)
print('Cricket player: ', is_cricket)
```
